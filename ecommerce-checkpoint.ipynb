```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

customers = pd.read_csv('Customers.csv')
products = pd.read_csv('Products.csv')
transactions = pd.read_csv('Transactions.csv')

print(customers.head())
print(products.head())
print(transactions.head())

print(customers.isnull().sum())
print(products.isnull().sum())
print(transactions.isnull().sum())

customers.dropna(inplace=True)
products.dropna(inplace=True)
transactions.dropna(inplace=True)

merged = pd.merge(transactions, customers, on='CustomerID', how='inner')
merged = pd.merge(merged, products, on='ProductID', how='inner')

print(merged.describe())

sns.histplot(merged['Quantity'], kde=True)
plt.title('Distribution of Quantity')
plt.show()

product_sales = merged.groupby('ProductName')['TotalValue'].sum().sort_values(ascending=False)
product_sales.head(10).plot(kind='bar', color='c')
plt.title('Top 10 Products by Total Sales')
plt.show()

customer_sales = merged.groupby('CustomerID')['TotalValue'].sum().sort_values(ascending=False)
top_10_customers = customer_sales.head(10)
print("Top 10 Customers by Total Sales:", top_10_customers)

region_sales = merged.groupby('Region')['Quantity'].sum()
region_sales.plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Product Sales by Region')
plt.show()

avg_spend = customer_sales.mean()
print(f"Average Spend per Customer: ${avg_spend:.2f}")

merged['TransactionDate'] = pd.to_datetime(merged['TransactionDate'])
sales_trend = merged.groupby(merged['TransactionDate'].dt.to_period('M'))['TotalValue'].sum()
sales_trend.plot(kind='line', color='b')
plt.title('Sales Trend Over Time')
plt.show()

top_10_profitable_products = product_sales.head(10)
print("Top 10 Most Profitable Products:", top_10_profitable_products)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
import seaborn as sns

# Prepare customer profiles for clustering
customer_profiles_for_clustering = merged.groupby('CustomerID').agg({
    'TotalValue': 'sum',
    'Quantity': 'sum',
    'Price': 'mean'
}).reset_index()

scaler = StandardScaler()
normalized_data = scaler.fit_transform(customer_profiles_for_clustering[['TotalValue', 'Quantity', 'Price']])

kmeans = KMeans(n_clusters=5, random_state=42)
customer_profiles_for_clustering['Cluster'] = kmeans.fit_predict(normalized_data)

db_index = davies_bouldin_score(normalized_data, customer_profiles_for_clustering['Cluster'])
print(f"Davies-Bouldin Index: {db_index}")

sns.scatterplot(data=customer_profiles_for_clustering, x='TotalValue', y='Quantity', hue='Cluster', palette='Set1')
plt.title('Customer Segmentation (Clustering)')
plt.show()

customer_profiles_for_clustering.to_csv('Customer_Segmentation.csv', index=False)
```
